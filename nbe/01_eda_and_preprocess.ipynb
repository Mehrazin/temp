```python
# Notebook 01: eda_and_preprocess
from sampling_framework import SamplingFramework

sf = SamplingFramework(spark)
input_table = "synthetic_customers_raw"
output_table = "processed_customers"

df = spark.table(input_table)

# 1. EDA Hook
numeric_cols = ["balance", "n_web_logins", "n_mobile_logins", "n_offers_year"]
summary, null_pcts = sf.get_summary_stats(df, numeric_cols)
summary.show()
print(f"Null distribution: {null_pcts}")

# 2. Preprocess
config = {
    'categorical_cols': ['visa_ind'],
    'continuous_cols': ['balance'],
    'discrete_numeric_cols': ['n_web_logins', 'n_mobile_logins', 'n_offers_year']
}

processed_df = sf.preprocess_data(df, config)

# 3. Save
processed_df.write.format("delta").mode("overwrite").saveAsTable(output_table)
print(f"Preprocessed data saved to {output_table}")
```
