```python
# Notebook 02: random_sampling
from sampling_framework import SamplingFramework

sf = SamplingFramework(spark)
df = spark.table("processed_customers")

# Create 90/10 Split
test_df, ctrl_df = sf.random_sample(df, id_col="cust_ID", test_fraction=0.9)

# Save
test_df.write.format("delta").mode("overwrite").saveAsTable("random_test_group")
ctrl_df.write.format("delta").mode("overwrite").saveAsTable("random_ctrl_group")

print(f"Random Test size: {test_df.count()}, Control size: {ctrl_df.count()}")
```
