```python
# Notebook 00: generate_synthetic_data
import numpy as np
from pyspark.sql import Row

n_rows = 10000 
cust_ids = list(range(1000000, 1000000 + n_rows))
balance = np.random.lognormal(mean=8, sigma=1.5, size=n_rows)
web_logins = np.random.poisson(lam=0.8, size=n_rows)
mobile_logins = np.random.poisson(lam=0.5, size=n_rows)  # Similar distribution to web_logins
visa_ind = np.random.choice([1, 0, None], size=n_rows, p=[0.38, 0.54, 0.08])
offers = np.random.randint(1, 50, size=n_rows)

data = []
for i in range(n_rows):
    data.append(Row(
        cust_ID=cust_ids[i], 
        balance=float(balance[i]), 
        n_web_logins=int(web_logins[i]),
        n_mobile_logins=int(mobile_logins[i]),
        visa_ind=int(visa_ind[i]) if visa_ind[i] is not None else None, 
        n_offers_year=int(offers[i])
    ))

df = spark.createDataFrame(data)
df.write.format("delta").mode("overwrite").saveAsTable("synthetic_customers_raw")
display(df.limit(10))
```
